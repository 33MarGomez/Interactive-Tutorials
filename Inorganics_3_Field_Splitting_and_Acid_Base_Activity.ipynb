There is code included around ranking contributions to the final MO, using the AO coefficients. Many of those coefficients correspond to high momentum orbitals, centred on atoms that could not sustain them. This is a product of the b97-d method, and of the def2 basis set. The latter adds $+1 \ell$ contributions, but the method uses a series of solutions - functions of functions or functionals - to properly adjust the d-orbital orientations.
