# Introduction

$T_1$ and $T_2$ relaxation times are vital aspects of ubiquitous NMR techniques including NOE, pore analysis, and chemical exchange effects. These critically require a motion-decoupled moment to be carried by solution to recompose the Larmor frequency, $\omega_0$. This causes the characteristic motion of a molecule in solution to carry information according to the statistics of its carrying $\omega_0$ in its freedom of motion. Those statistics are represented by Spectral Density, $J(\omega)$, which depicts the odds that any spinning element carries a particular angular frequency.

In this work, $\tau_C$ is the root mean square deflection of the molecule by one radian, 57 degrees. Such an arbitrary description is obviated by just generally calling $\tau_c$ the unique, or characteristic, amount of time it takes the molecule to act in its own characteristic way, ie undergoing its characteristic series of rotational motions. In this way, an instantaneous $J(\omega)$ description is non-descriptive because the molecule is frozen in place. $J(\omega)$ is more aptly described as a time-dependent pheonomenon of any rotational element carrying the motion at all in a period of time, instead of being spatially descriptive. This is called the random-fields approximation. This tutorial graphs $\omega_0J(\omega_0)$ as a correlation phenomenon, like the value carried in $\tau_c\omega_0$ if $\tau_c$ is $\alpha$ occurances of $\omega_0$ in characteristic lapses of time.

$\tau_c^{-1} = \omega_c$

$\tau_c = \frac{\alpha}{\omega_0}$

$\alpha = \frac{\omega_0}{\omega_c}$

Throughout this kind of work, $J(0)$ and $J(2\omega_0)$ are seen to be critical solutions. $J(0)$ simply means that two molecules are spinning synchronously, and because their Larmor frequencies are individually decoupled from their motion, they can exchange labels. This is best imagined as orthogonal spokes on a wheel, with arrows pointing up and down regardless of the spinning. It can also describe two arrows turning in sequence about a common point, hence the $J(0)$ importance in finding chemical proximity.

The $J(2\omega_0)$ necessity is not more complicated. It means that where twice the speed of the Larmor frequency is happening, one moment is seeing the Larmor frequency 'cut' its coil in its own cone. This causes the double spin exchange with no loss of momentum (emission). The period-angular velocity relationship can be used to demonstrate the difference in periods is also the difference in angular velocities.

$T_{\omega_0} = \frac{2\pi}{\omega_0}$

$Δω = 2\omega_0 - \omega_0 = 2T_{\omega_0}^{-1} - T_{\omega_0}^{-1}$

$\Delta \omega = T_{\omega_0}^{-1} = \omega_0$


This basic trigonometric principle is most cleverly demonstrated as a solution to an equation demonstrating twice coincident $T_{\omega_0}$ solutions arising from $T_{2\omega_0}$ inclusion.

$T_{2\omega_0} = \frac{2\pi}{2\omega_0}$

$2T_{2\omega_0} = \frac{2\pi}{\omega_0} = T_{\omega_0}$

$0 = 2T_{2\omega_0} - T_{\omega_0}$

$0 = \frac{2}{2}(T_{\omega_0}) - T_{\omega_0} = \frac{1}{2}T_{\omega_0} + \frac{1}{2}T_{\omega_0} - T_{\omega_0} $

$0 = \{ nT_{\omega_0} + nT_{\omega_0} - 2nT_{\omega_0} | n \in ℕ \}$

# Visualizing Spectral Density

In general, we can call $\tau_c$ an $\alpha$ multiple of Larmor frequency,

$τ_c = α / \omega_0$

and every frequency an x multiple of Larmor frequency,

$\omega = x * \omega_0$

The frequency can also be called a multiple of Larmor frequency. Together, this leads to a plot over $1 + x^2$ with $\tau_c$ modifying the numerator. x is the multiple of $\omega_0$ indexing the Spectral Density at that $\omega$, making $\tau_c$ relationships parametric.

$J(\omega_0) * \omega_0 = \frac{\tau_c}{1+\omega^2\tau_c^2}$

$J(\omega_0) * \omega_0 = \{ \frac{1}{\omega_0} \} \frac{2\alpha}{1+(\omega_0^2x^2)(\alpha^2\omega_0^{-2})}$

$J(\omega_0) * \omega_0 = \frac{2\alpha}{1+\alpha^2x^2}$


```python
import numpy as np
import matplotlib.pyplot as plt

xaxis = np.arange(0, 10, 0.01)
yaxis = 2*4 / ((1 + (4**2)*(xaxis**2)))
y2axis = 2*1 / ((1 + (1**2)*(xaxis**2)))
y3axis = 2*0.25 / ((1 + (0.25**2)*(xaxis**2)))

fig, ax = plt.subplots()
ax.plot(xaxis,yaxis,label=r"TauC = 4")
ax.plot(xaxis,y2axis,label=r"TauC = 1")
ax.plot(xaxis,y3axis, label = r"TauC = 0.25")
plt.ylabel("Spectral Density Scaled Larmor Frequency")
plt.xlabel(r"$\omega$ in x multiples of $\omega_0$")
ax.legend(fontsize=14)
plt.show()
```

# $T_1$ Dependence on Spectral Density

$T_1$, Spin-Lattice Relaxation time, encompasses the random fields synchronizing with $\omega_0$. This depends on relative rotation randomly oscillating at the Larmor Frequency, causing the rotation-decoupled moment to cause a transition.

$\frac{1}{T_1} = \gamma^2 \langle B_{loc}^2 \rangle J(\omega_0)$

Reducing over gyromagnetic moment and mean-square of local fluctuating field,

$T_1 = \frac{1}{J(\omega_0)}$

$T_1 / \omega_0 = \frac{1+α^2}{2*α}$

$T_1 \propto \frac{1+α^2}{2*α}$

By,

$T_1 = \frac{1+\omega_0^2\tau_c^2}{2*\tau_c}$

$T_1 = \omega_0\frac{1+\omega_0^2\alpha^2\omega_0^{-2}}{2*\alpha}$

In rapid tumbling, correlation time is much too small for Larmor frequency, causing the inequality:

$\omega_0 * \tau_c << 1$

Therefore, correlation time must be slowed to decrease relaxation time so the environment can contain some element of the Larmor frequency.

In slow tumbling, the correlation time contains many multiples of the angular frequency because it's rotating slowly.

$\omega_0 * \tau_c >> 1$

This causes relaxation time to become faster as correlation time is decreased because the molecule needs to speed up to match Larmor frequency.

Shown below will be the α number of times the Larmor frequency occurs, such that $T_1$ can be plotted in $\frac{1}{J(\omega)}$ as α happenings per correlation time. Please note that this is occurance of Larmor frequency not reduced by $2\pi$ or ħ, and only depicts $\propto \omega_0$, not an NMR frequency.

```python
import numpy as np
import matplotlib.pyplot as plt

xaxis = np.arange(0.1, 10, 0.01)
yaxis = ((xaxis**2) + 1) / (2 * xaxis)


fig, ax = plt.subplots()
ax.plot(xaxis,yaxis)

plt.xlabel(r"$\alpha$ occurances of $\omega_0$ in $\tau_c$")
plt.ylabel(r"$T_1$ proportionality")
plt.show()
```

# Predicting NOE Intensity

NOE intensity, $\eta$, will generally follow in sign according to cross relaxation processes between $\alpha$$\alpha$ ↔ $\beta$$\beta$ and $\alpha \beta \leftrightarrow \beta \alpha$, depicted $W_2^{AX}$ and $W_0^{AX}$ respectively. These are known to exhibit proportionality:

$W_2^{AX} \propto J(2\omega_0)$

$W_0^{AX} \propto J(0)$

The sign of NOE signal enhancement matches the sign of subtraction of the mechanisms.

$sign(\eta) = W_2^{AX} - W_0^{AX}$

Minding that $\eta$ is a depiction of intensity from an unstimulated intensity,

$\eta = \frac{i-i_0}{i_0}$

Depicted below are several NOE predictions with different $\tau_c$. As one might guess, using the y-intercept of the $J(\omega_0)\omega_0$ to $ \alpha\omega_0$ plot and subtracting the x = 2 y-point in that plot gives the sign of NOE enhancement. This can be readily demonstrated:

$J(0) * \omega_0 = \frac{2\alpha}{1+\alpha^2(0)} = 2\alpha$

$J(2\omega_0) * \omega_0 =  \frac{2\alpha}{1+2^2\alpha^2} = \frac{2\alpha}{1+(2\omega_0)^2(\alpha^2\omega_0^{-2})}$

```python
import numpy as np
import matplotlib.pyplot as plt

xaxis = np.arange(0, 10, 0.01)
yaxis = 2*4 / ((1 + (4**2)*(xaxis**2)))
y2axis = 2*1 / ((1 + (1**2)*(xaxis**2)))
y3axis = 2*0.25 / ((1 + (0.25**2)*(xaxis**2)))

fig, ax = plt.subplots()
ax.plot(xaxis,yaxis,label=r"TauC = 4")
ax.plot(xaxis,y2axis,label=r"TauC = 1")
ax.plot(xaxis,y3axis, label = r"TauC = 0.25")
ax.axline((2, 4), slope=1000, color="black", linestyle=(0, (5, 5)))
plt.ylabel("Spectral Density Scaled Larmor Frequency")
plt.xlabel(r"$\omega$ in x multiples of $\omega_0$")
ax.legend(fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = 2
p = ((2*x) /(1+(4*(x**2)))) - 2*x
plt.scatter(x,p)

x = 4
p = 2*x/(1+(4*(x**2))) - 2*x
plt.scatter(x,p)

x = 1
p = 2*x/(1+(4*(x**2))) - 2*x
plt.scatter(x,p)

x = 0.5
p = 2*x/(1+(4*(x**2))) - 2*x
plt.scatter(x,p)

x = 0.2
p = 2*x/(1+(4*(x**2))) - 2*x
plt.scatter(x,p)

x = 0.01
p = ((2*x) /(1+(4*(x**2)))) - 2*x
plt.scatter(x,p)

plt.ylabel(r"NOE Intensity Proportionality")
plt.xlabel(r"$\alpha$ multiples of $\omega_0$ in $\tau_c$")
plt.show()
```

The positive domain appears to be missing. This is expected, that domain is actually dependent on modelling $W_0^{AX}$ and $W_2^{AX}$ and not on spectral density in rapid tumblers. This is standard in very tiny quantum mechanical quantities, particularly since NOE does use an $r^{-6}$ net dependence. In this absence, the desireable outcome for a positive signal would have been a saturated $\beta\beta$ level replenishing $\alpha\alpha$ through $W_2^{AX}$ to undergo a (freshly saturated) positive signal.

The negative domain is linearly negative. The Larmor frequency must scale by a factor of $r^6$ to keep up, making NOE very useful for probing protein and stereochemical proximities. Equally, in a slow tumbling molecule, saturating a spin then participating in population inversion will cause those spins to equalize via dipolar transfer $\alpha\beta ↔ \beta\alpha$ transfer, $J(0)$. The desired transition will cause negative signals to a mixed state, which $W_0^{AX}$ transfers to undergo a second negative signal.

# $T_2$ dependences on Spectral Density

Spin-Spin Relaxation occurs in liquids as peak narrowing caused by an average molecular environment. Tumbling rate $\tau_c^{-1}$ approaches the product of magnetogyric ratio reduced over dipolar radius, $\tau_c^{-1} → \frac{\gamma_A \gamma_X}{r^3} $, averaging the environment. Note it is not productive to think about this value as composing $\hbar \gamma_A \gamma_X$ because it will generally be a range of motions decaying over the cubic radius which is subsequently squared for $r^{-6}$. It is taken in a random fields mechanism to occur through a $J(\omega_0)$-dependent $T_1$ term and a $J(0)$ term motional narrowing term. $J(0)$ will narrow the peak at low $\tau_c$ by averaging chemical environments and preventing transfer, per reciprocal appearance increasing $T_2$. Similarly, $J(0)$ will create labile transfer at $\alpha$ >> 1 in the slow motion regime.

$\frac{1}{T_2} =\{\frac{1}{2} \gamma^2 \langle B_{loc}^2 \rangle \} \{ J(\omega_0) + J(0) \}$

$T_2 \propto \{ J(\omega_0) + J(0) \}^{-1}$

This can be readily demonstrated:

$J(0) * \omega_0 = \frac{2\alpha}{1+\alpha^2(0)} = 2\alpha$

$J(2\omega_0) * \omega_0 =  \frac{2\alpha}{1+2^2\alpha^2}$

The reciprocal Spectral Density sum has been plotted below. This involved taking y-intercepts of the $J(\omega)$ graphs seen above ie 2 $\tau_c$, which are added to the $J(2\omega_0)$ calculation as described.

```python
import numpy as np
import matplotlib.pyplot as plt

x = 2
p = (2 * x + 2*x/(1+4*x**2))**-1
plt.scatter(2,p)

x = 4
p = (2 * x + 2*x/(1+4*x**2))**-1
plt.scatter(4,p)

x = 1
p = (2 * x + 2*x/(1+4*x**2))**-1
plt.scatter(1,p)

x = 0.5
p = (2 * x + 2*x/(1+4*x**2))**-1
plt.scatter(0.5,p)

x = 0.2
p = (2 * x + 2*x/(1+4*x**2))**-1
plt.scatter(0.2,p)


plt.ylabel(r"$T_2$ Proportionality")
plt.xlabel(r"$\alpha$ multiples of $\omega_0$ in $\tau_c$")
plt.show()
```

This reciprocal relationship is expected. Where $\alpha << 1$, the result is an infinitely decayed $\alpha^2$ term:

$T_2 \propto \Bigl\{2\alpha + \frac{2\alpha}{1+4\alpha^2} \Bigl\}^{-1}$

$T_2 \propto \Bigl\{2\alpha + \frac{2\alpha}{1+0} \Bigl\}^{-1}$

$T_2 \propto \{ 4\alpha \}^{-1}$

$T_2 \propto \frac{1}{4\alpha} = \infty$

Therefore, in rapid tumbling molecules such that $\omega_0\alpha\omega_0^{-1} = \omega_0\tau_c << 1$ the molecules are tumbling much too fast to transfer state, and the signal is long lived. $J(0)$ and $J(2\omega_0)$ are included by this solution because they're equally impotent. This graphical solution is rationalized by the long, flat curves created by small $\tau_c$ in $J(\omega)\omega_0$ versus $\omega$.

Where $\alpha >> 1$, the $J(2\omega_0)$ term decays to 0:

$T_2 \propto \{ 2\alpha + 0 \}^{-1}$

$T_2 \propto \frac{1}{\alpha}$

$T_2 → 0$

Only the reciprocal time contribution of the $J(0)$ term is considered in slow tumblers. One might consider this similarly to a solid experiencing coherent state transfer, particularly when examining the tall, sharp curves created by high $\tau_c$ in plot $J(\omega)\omega_0$ to $\omega$. These tall intercepts all represent coherent dipolar transfer in a slow motion domain.
